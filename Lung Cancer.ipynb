<p style="text-align:justify; ">
    
Lung cancer is a type of cancer that begins in the lungs. Lungs are two spongy organs in our chest that take in oxygen when we inhale and release carbon dioxide when we exhale.
Lung cancer is the leading cause of cancer deaths worldwide.<br>
<br>
People who smoke have the greatest risk of lung cancer, though lung cancer can also occur in people who have never smoked. The risk of lung cancer increases with the length of time and number of cigarettes someone's smoked. If someone quits smoking, even after smoking for many years, he/she can significantly reduce the chances of developing lung cancer.

# <p style="background-color:purple; font-family:calibri; color:white; font-size:170%; text-align:center; border-radius:45px 50px;">Step 5 |  Data Preprocessing</p>

```python
# Replace Cat to Numeric for all column

## GENDER 
df['GENDER'].replace('Male',0, inplace=True)
df['GENDER'].replace('Female',1, inplace=True)

## SMOKING
df['SMOKING'].replace('NO',0, inplace=True)
df['SMOKING'].replace('YES',1, inplace=True)

## YELLOW_FINGERS
df['YELLOW_FINGERS'].replace('NO',0, inplace=True)
df['YELLOW_FINGERS'].replace('YES',1, inplace=True)

## ANXIETY
df['ANXIETY'].replace('NO',0, inplace=True)
df['ANXIETY'].replace('YES',1, inplace=True)

## PEER_PRESSURE
df['PEER_PRESSURE'].replace('NO',0, inplace=True)
df['PEER_PRESSURE'].replace('YES',1, inplace=True)

## CHRONIC DISEASE
df['CHRONIC_DISEASE'].replace('NO',0, inplace=True)
df['CHRONIC_DISEASE'].replace('YES',1, inplace=True)

## FATIGUE
df['FATIGUE'].replace('NO',0, inplace=True)
df['FATIGUE'].replace('YES',1, inplace=True)

## ALLERGY
df['ALLERGY'].replace('NO',0, inplace=True)
df['ALLERGY'].replace('YES',1, inplace=True)

## WHEEZING
df['WHEEZING'].replace('NO',0, inplace=True)
df['WHEEZING'].replace('YES',1, inplace=True)

## ALCOHOL CONSUMING
df['ALCOHOL_CONSUMING'].replace('NO',0, inplace=True)
df['ALCOHOL_CONSUMING'].replace('YES',1, inplace=True)

## COUGHING
df['COUGHING'].replace('NO',0, inplace=True)
df['COUGHING'].replace('YES',1, inplace=True)

## SHORTNESS OF BREATH
df['SHORTNESS_OF_BREATH'].replace('NO',0, inplace=True)
df['SHORTNESS_OF_BREATH'].replace('YES',1, inplace=True)

## SWALLOWING DIFFICULTY
df['SWALLOWING_DIFFICULTY'].replace('NO',0, inplace=True)
df['SWALLOWING_DIFFICULTY'].replace('YES',1, inplace=True)

## CHEST PAIN
df['CHEST_PAIN'].replace('NO',0, inplace=True)
df['CHEST_PAIN'].replace('YES',1, inplace=True)

## LUNG_CANCER
df['LUNG_CANCER'].replace('NO',0, inplace=True)
df['LUNG_CANCER'].replace('YES',1, inplace=True)
```

```python
# Create the correlation heatmap with annotations
plt.figure(figsize=(15, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")  # annot=True to show numbers, fmt=".2f" to format to 2 decimal places
plt.title('Correlation Heatmap of Lung Cancer Features')
plt.show()
```

```python
# storing the Dependent Variables in X and Independent Variable in Y
X=df.drop(['LUNG_CANCER'],axis=1)
y=df['LUNG_CANCER']
```

```python
X
```

```python
y
```

```python
# split the data into training and testing sets using the train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  random_state=2)
```

<a id="import"></a>
# <p style="background-color:purple; font-family:calibri; color:white; font-size:170%; text-align:center; border-radius:45px 50px;">Step 6 |  Model Building</p>

<a id="import"></a>
# <p style="background-color:blue; font-family:calibri; color:white; font-size:170%; text-align:center; border-radius:45px 50px;">Step 6.1 |  LogisticRegression</p>

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
X_test_prediction = model.predict(X_test)
acc_1 = accuracy_score(X_test_prediction, y_test)
print('Accuracy score on Test Data : ', acc_1)
```

```python
conf_matrix = confusion_matrix(y_test, X_test_prediction)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
class_report = classification_report(y_test, X_test_prediction)
print("Classification Report:")
print(class_report)
```

<a id="import"></a>
